#### Import Needed Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("fstreet.csv")
data.head()
```

```python
data.state_code.value_counts()
```

```python
cal_data = data.loc[data['state_code'] == 'CA']
```

```python
cal_data.columns
```

```python
cal_data.head()
```

```python
cal_data.isnull().sum()
```

```python
## Filter for the most recent location of the
## the hurricane
cal_data.flood_risk_level.value_counts()
```

```python
cali_data = cal_data[['geo_name', 'state_code', 'property_count', 'structure_count', 'flood_risk_level']]
cali_data.head()
```

```python
risk_mapping = {
    'none': 0,
    'minimal': 1,
    'minor': 2, 
    'moderate': 3,
    'major': 4,
    'severe': 5,
    'extreme': 6
}
```

```python
cali_data['flood_risk_score'] = cali_data['flood_risk_level'].map(risk_mapping)
```

```python
cali_data.head()
```

```python
## Heat Map
contingency_table = pd.crosstab(cali_data['geo_name'], cali_data['flood_risk_level'])

plt.figure(figsize=(10, 8))
sns.heatmap(contingency_table, annot=True, fmt=".1f", cmap='Blues')
plt.title("Flood Risk Score by County in California")
plt.xlabel("Flood Risk Score")
plt.ylabel('County')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.scatterplot(data=cali_data, x = 'property_count', y='flood_risk_score', hue='Count', s=100)
plt.title('Flood Risk Score vs. Property Count')
plt.xlabel('')
```
